```python
import os

from random import shuffle

from math import floor

import shutil

#https://pynative.com/python-copy-files-and-directories/
#https://stackoverflow.com/questions/42471570/how-to-split-documents-into-training-set-and-test-set
```

```python

def get_file_list_from_dir(datadir):
    all_files = os.listdir(os.path.abspath(datadir))
    data_files = list(filter(lambda file: file.endswith('.npz'), all_files))
    return data_files

def randomise_files(file_list):
    shuffle(file_list)

def get_training_and_testing_sets(file_list):
    split = 0.5
    split_index = floor(len(file_list) * split)
    training = file_list[:split_index]
    testing = file_list[split_index:]
    return training, testing
```

```python
file_list=get_file_list_from_dir('/Users/lucascurtin/Desktop/QGNN Repos/HepTrkX-quantum/data/tuysuz_100_events/all')
```

```python
randomise_files(file_list)
```

```python
training,testing=get_training_and_testing_sets(file_list)
```

```python
source ='all'
training_destination = 'train'

for f in training:
    src_path = os.path.join(source, f)
    dst_path = os.path.join(training_destination, f)
    shutil.move(src_path, dst_path)
```

```python
source_folder = r"all/"
train_folder = r"train/"

# fetch all files
for file_name in training:
    # construct full file path
    source = source_folder + file_name
    destination = train_folder + file_name
    print(source)
    print(destination)
    # copy only files
    if os.path.isfile(source):
        shutil.copy(source, destination)
        #print('copied', file_name)
```

```python
source_folder = r"all/"
valid_folder = r"valid/"

# fetch all files
for file_name in testing:
    # construct full file path
    source = source_folder + file_name
    destination = valid_folder + file_name
    print(source)
    print(destination)
    # copy only files
    if os.path.isfile(source):
        shutil.copy(source, destination)
        #print('copied', file_name)
```
